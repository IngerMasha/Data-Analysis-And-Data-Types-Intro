```python
import pandas as pd
import matplotlib.pyplot as plt
iris_data = pd.read_csv('/content/sample_data/iris.csv')
mean_sepal_length = iris_data['sepal_length'].mean()
median_sepal_length = iris_data['sepal_length'].median()
mode_sepal_length = iris_data['sepal_length'].mode()[0]

```

```python
print("The average length of the chapel:", mean_sepal_length)
print("Mediana of the length of the cup:", median_sepal_length)
print("Fashion for the length of the cups:", mode_sepal_length)
```

```python
plt.hist(iris_data['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('The length of the cups (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of the length of the cups')
plt.show()
```
